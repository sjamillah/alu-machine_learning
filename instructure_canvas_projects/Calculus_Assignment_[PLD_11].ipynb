# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
# Denys
#TO DO: Create a function that does a derivative for any function
import sympy as sp

def dy_dx(fx):
  #find the derivative of f(x)
  x = sp.symbols('x')

  f = sp.sympify(fx)
  derivative = sp.diff(f, x)

  return derivative
# How it is used

input_function = 'x**4 - 8*x**2 + 12'  #it's possible to enter any function as string
result = dy_dx(input_function)

print(f"The derivative of {input_function} is: {result}")
````

#Task 3: Plot a graph of the Equation must have 4 points (2 minimas and 2 maximas)

````python
#Ssozi
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np

# Equation created in task 2
def f(x):
    return x**4 - 8*x**2 + 12

# Derivative of the function
def df(x):
  return 4*x**3 - 16*x

# Find the maxima and minima values
x = sp.symbols('x')
f_expression = f(x)
df = sp.diff(f_expression, x)

# Find the critical points (minima and maxima)
critical_points = sp.solve(df, x)

# Convert critical points to float for plotting
critical_points = [float(cp) for cp in critical_points]

# Convert symbolic function and derivative to numerical functions for plotting
f_lambdified = sp.lambdify(x, f_expression, 'numpy')
df_lambdified = sp.lambdify(x, df, 'numpy')

# Create a visualization of the quadratic/algebraic equation
x_values = np.linspace(-3, 3, 400)
y_values = f_lambdified(x_values)
dy_values = df_lambdified(x_values)

# plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$f(x) = x**4 - 8*x**2 + 12$')
plt.plot(x_values, dy_values, label=r"$f'(x) = 4*x**3 - 16*x$")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# plot the critical points(minima and maxima)
plt.scatter(critical_points, [f_lambdified(cp) for cp in critical_points], color='green', label='Critical Points', zorder=5)

# Add labels, title, and grid
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function with Maxima and Minima')
plt.legend()
plt.grid(True)
plt.show()
````

# Task 4: Create separate array of Minimas and maximas

````python
#Sifa
import sympy as sp
import numpy as np

# Define the function from Task 2
def f(x):
    return x**4 - 8*x**2 + 12
# Symbolic computation to find critical points
x = sp.symbols('x')
f_sym = f(x)
df = sp.diff(f_sym, x)  # First derivative
d2f = sp.diff(df, x)    # Second derivative

# Solve for critical points
critical_points = sp.solve(df, x)

# Separate arrays for the minima and maxima
arr_minimas = []
arr_maximas = []

# Classify critical points as minima or maxima using the second derivative test
for cp in critical_points:
  if sp.im(cp) == 0:  # Check if the critical point is real
    second_derivative_at_cp = d2f.subs(x, cp).evalf()
    if float (second_derivative_at_cp) > 0:
        arr_minimas.append(cp)
    elif float (second_derivative_at_cp) < 0:
        arr_maximas.append(cp)

# Print the arrays of minima and maxima
print(f"The array of minima: {arr_minimas}")
print(f"The array of maxima: {arr_maximas}")
````
