CO2 emissions histogram

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
plt.hist(dataset['co2'], bins=10, edgecolor='black')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.title('Distribution of CO2 Emissions')
plt.show()
```

Playing around with the bin sizes

```python
# Play around with different bin sizes for CO2 emissions
plt.figure(figsize=(10, 6))
plt.hist(dataset['co2'], bins=30, edgecolor='black')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.title('Distribution of CO2 Emissions with Smaller Bins')
plt.show()
```

```python
# Select columns with numerical data types
numeric_data = dataset.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the heatmap of the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.show()
```
